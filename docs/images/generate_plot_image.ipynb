```python
import covid19pandas as cod
global_df = cod.get_data_jhu()

us_df = cod.select_regions(data=global_df, region_col="Country/Region", regions="US")
meaned = cod.calc_x_day_rolling_mean(data=us_df, data_cols="cases", region_cols="Country/Region", x=5)
with_daily = cod.calc_daily_change(data=meaned, data_cols="mean_cases", region_cols="Country/Region")

fig, ax = cod.plot_lines_two_y(data=with_daily, x_col="date", y1_col="daily_mean_cases", y2_col="mean_cases", title="Date vs daily new cases and cumulative cases for U.S.", x_lab="Date", y1_lab="Daily new cases (5 day rolling mean)", y2_lab="Cumulative cases (5 day rolling mean)", legend_loc="center left", dimensions=(15, 8), seaborn_style="whitegrid")
```
